### Loading folder having 11 images and appending all in a list

```python
import glob
from skimage.io import imread, imshow

imgset = []
for img in glob.glob("C:/Users/Vivek/Documents/dogs/*.jpg"):
    n= imread(img)
    imgset.append(n)
```

### Converting all images into grayscale

```python
from skimage.color import rgb2gray # converting to grayscale so that our image has only one color channel

gray_img_set = []
for i in imgset:
    grayimg = rgb2gray(i)
    gray_img_set.append(grayimg)
```

### Extracting image pixel features and storing them all in a numpy array

```python
for image in imgset:
    feature_matrix_image = np.zeros((image.shape)) 
    feature_matrix_image
    for i in range(0,image.shape[0]):
        for j in range(0,image.shape[1]):
            feature_matrix_image[i][j] = ((int(image[i,j,0]) + int(image[i,j,1]) + int(image[i,j,2]))/3)
type(feature_matrix_image)
```
